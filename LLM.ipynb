```python
import csv

# Define the data to be written to the CSV file
csv_content = [
    ['Question', "NuminaMath_7B_TIR","Answers"],
]





# Example: Link to your LM Studio
# Point to the local server
from openai import OpenAI
client = OpenAI(base_url="http://localhost:1234/v1", api_key="not-needed")



import json
a = 0
with open('test_data.jsonl', 'r', encoding="utf-8") as f:
    for line in f:
        data = json.loads(line)
        # upload a math problem
        tmp_question = data["question"]


        completion_NuminaMath = client.chat.completions.create(
        model="lmstudio-community/reach-vb/NuminaMath-7B-TIR-Q8_0-GGUF/numinamath-7b-tir-q8_0.gguf", # this field is currently unused
        messages=[
        {"role": "system", "content": tmp_question},
   
          ],
          temperature=0.7,
)

        
        csv_content.append([data["question"],completion_NuminaMath.choices[0].message.content,data["answer"]])
        print(completion_NuminaMath.choices[0].message.content)
        a += 1
        print(a)
        # how many problem a batch you want
        if a ==1000:
            break



# Specify the file name
filename = 'GSM8K_test_completion_NuminaMath_7B_TIR.csv'

# Write data to the CSV file with a pipe delimiter
with open(filename, 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile, delimiter=',')
    csvwriter.writerows(csv_content)
print(f"Data has been written to {filename}")
```
